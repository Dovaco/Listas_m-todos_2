## Exercício 1:

```python
import numpy as np
import statistics as st

dados = np.array([0.90, 1.42, 1.30, 1.55, 1.63, 1.32, 1.35, 1.47, 1.95, 1.66, 1.96, 1.47,
                  1.92, 1.35, 1.05, 1.85, 1.74, 1.65, 1.78, 1.71, 2.29, 1.82, 2.06, 2.14, 
                  1.27])

mean = np.mean(dados) 
median = np.median(dados) 
mode = st.mode(dados)
var = np.var(dados)
std = np.std(dados)
coef_var = (std / mean) * 100

print('Resultados:'
'\nMédia = %.2f'
     '\nMediana = %.2f'
     '\nModa = %.2f'
     '\nVariância = %.2f'
     '\nDesvio padrão = %.2f'
     '\nCoeficiente de variação = %.2f%%'
      %(mean, median, mode, var, std, coef_var))
```

## Exercício 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

bins = np.arange(0.8, 2.4, 0.2)

hist = np.histogram(dados, bins)

plt.hist(dados, bins='auto', color='khaki',edgecolor='black')
plt.title('Histograma')

#normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2, label=f'Normal (μ={mean:.2f}, σ={std:.2f})')
plt.title('Distribuição normal')
plt.legend()
plt.show()
```

## Exercício 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

dados1 = np.array([
    29.65, 28.55, 28.65, 30.15, 29.35, 29.75, 29.25, 30.65, 28.15,
    29.85, 29.05, 30.25, 30.85, 28.75, 29.65, 30.45, 29.15, 30.45,
    33.65, 29.35, 29.75, 31.25, 29.45, 30.15, 29.65, 30.55, 29.65, 29.25
])

mean1 = np.mean(dados1)
std1 = np.std(dados1)

#histograma
plt.hist(dados1, bins='auto', density=True, alpha=0.6, color='cadetblue', edgecolor='black')

#normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean1, std1)
plt.plot(x, p, 'k', linewidth=2, label=f'Normal (μ={mean1:.2f}, σ={std1:.2f})')
plt.title('Distribuição normal')
plt.legend()
plt.show()
```

```python
#intervalo que engloba 68% dos valores:

# X = Z * σ + μ

z_min, z_max = -0.99, 0.99

x_sup = (z_max * std1) + mean1
x_inf = ((z_min) * std1) + mean1

plt.figure(figsize=(10,6))
#curva normal padrão
z = np.linspace(-2, 2, 1000)
y = norm.pdf(z)

plt.plot(z, y, 'black', linewidth=1.5, label='Distribuição normal padrão')

#destacando o intervalo de 68%
z_int = np.linspace(z_min, z_max, 100)
y_int = norm.pdf(z_int)

plt.fill_between(z_int, y_int, color='plum', alpha=0.5, label='Intervalo -0.99 < z < 0.99')

#linhas de referência
plt.axvline(x=z_min, color='purple', linestyle='--')
plt.axvline(x=z_max, color='purple', linestyle='--')
plt.axvline(x=0, color='red', linestyle='-', label=f'Média = 0\nX min:{x_inf:.2f} X max:{x_sup:.2f}')


plt.title('Curva Normal Padrão com intervalo -0.99 < z < 0.99\n(que engloba 68% dos dados)')
plt.xlabel('Z')
plt.legend()
plt.grid(True, alpha=1)
plt.ylim(0, norm.pdf(0)*1.1)
plt.show()
```

## Exercício 5

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2, 4, 6, 7, 10, 11, 14, 17, 20])
y = np.array([4, 5, 6, 5, 8, 8, 6, 9, 12])
```

```python
def regres_linear(x : np.ndarray, y : np.ndarray, xscrib = 'X', yscrib = 'Y', title = 'Reta modelada de Y em função de X'):
    if len(x) != len(y):
        print('Erro')
        
    import numpy as np
    import matplotlib.pyplot as plt
    
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    xy = []
    x2 = []
    for i, j in zip(x, y):
        xy.append(i * j)
        x2.append(i ** 2)

    xy = np.array(xy)
    x2 = np.array(x2)
    
    #calculando b
    b = ((n * np.sum(xy)) - (np.sum(x) * np.sum(y))) / ((n * np.sum(x2)) - (np.sum(x) ** 2))
    #calculando a
    a = y_mean - (b * x_mean)

    #valores previstos
    y_prev = []
    for i in x:
        y_prev.append(a + (b * i))

    y_prev = np.array(y_prev)
    
    #calculando o erro
    e = 0
    for j, z in zip(y, y_prev):
        e_calc = ((j - z) ** 2)
        e += e_calc

    #coeficente de determinação
    sqt = 0
    for j in y:
        sqt_calc = (j - y_mean) ** 2
        sqt += sqt_calc
    r2 = 1 - (e / sqt)
    
    #plotando a reta:
    plt.figure(figsize=(10,6))
    plt.plot(x, y_prev, linewidth='2', color='black', label=f'y = {a:.2f} + {b:.2f} * x\n$R^2 = {r2:.2f}$')
    plt.scatter(x, y, color='red',label='Dados originais')
    plt.title(title)
    plt.xlabel(xscrib)
    plt.ylabel(yscrib)
    plt.grid(True)
    plt.legend()
    plt.show
```

```python
regres_linear(x, y)
```
